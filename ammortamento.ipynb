```python
import numpy as np
import pandas as pd
```

```python
PRESTITO = 40_000.0
ANNI = 5
TAN = 0.04
```

```python
# Ammortamento all'italiana
numero_rate = ANNI * 12
quota_capitale = round(PRESTITO / numero_rate)
capitale_residuo = PRESTITO

quote_capitale = []
quote_interessi = []

for i in range(numero_rate):
    quote_interessi.append(round(capitale_residuo * TAN / 12))
    capitale_residuo -= quota_capitale
    quote_capitale.append(quota_capitale)
    

df = pd.DataFrame({
    'quote_capitale': quote_capitale,
    'quote_interessi': quote_interessi
})
df['rate'] = df['quote_capitale'] + df['quote_interessi']
df['capitale_residuo'] = (PRESTITO - df['quote_capitale'].cumsum()).round()
```
