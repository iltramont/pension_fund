```python
import numpy as np
import pandas as pd
```

```python
PRESTITO = 200_000
ANNI = 30
TAN = 0.03
```

```python
# Ammortamento all'italiana
numero_rate = ANNI * 12
quota_capitale = round(PRESTITO / numero_rate, 2)

quote_capitale = [0]
quote_interessi = [0]
capitale_residuo = [PRESTITO]

for i in range(numero_rate):
    quote_interessi.append(round(capitale_residuo[i] * TAN / 12, 2))
    capitale_residuo.append(capitale_residuo[-1] - quota_capitale)
    quote_capitale.append(quota_capitale)

quote_capitale[-1] = capitale_residuo[-2]
capitale_residuo[-1] = 0
    

df = pd.DataFrame({
    'quote_capitale': quote_capitale,
    'quote_interessi': quote_interessi,
    'capitale_residuo': capitale_residuo
})
df['rate'] = df['quote_capitale'] + df['quote_interessi']
#df['capitale_residuo'] = (PRESTITO - df['quote_capitale'].cumsum()).round()
```
