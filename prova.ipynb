```python
import numpy as np
import pandas as pd
```

```python
p1 = 100
p2 = 150
p3 = 200
p4 = 250
```

```python
df = pd.DataFrame({
    'price': [p1, p2, p3, p4]
})
df['shifted'] = df['price'].shift(-1)
```

```python
df
```

```python
df['returns'] = (df['shifted'] - df['price']) / df['price']
df['log_ret'] = np.log(df['shifted'] / df['price'])
```

```python
df['cum_ret'] = (1 + df['returns']).cumprod() - 1
df['cum_log_ret'] = df['log_ret'].cumsum()
```
